## 3. Tratamento e preparação dos dados

 - coluna **'id_cliente'**


```python
dados = dados.drop('id_cliente', axis=1)
```

```python
dados.shape
```

- Tratamento de variáveis com categorias **Sim** e **Não**


```python
dados['cliente_ativo'] = dados['cliente_ativo'].map({'Sim' : 1, "Não" : 0})
dados['cliente_idoso'] = dados['cliente_idoso'].map({'Sim' : 1, "Não" : 0})
dados['cliente_parceiro'] = dados['cliente_parceiro'].map({'Sim' : 1, "Não" : 0})
dados['cliente_dependentes'] = dados['cliente_dependentes'].map({'Sim' : 1, "Não" : 0})
dados['linha_telefonica'] = dados['linha_telefonica'].map({'Sim' : 1, "Não" : 0})
dados['multiplas_linhas'] = dados['multiplas_linhas'].map({'Sim' : 1, "Não" : 0})
dados['seguranca_online'] = dados['seguranca_online'].map({'Sim' : 1, "Não" : 0})
dados['backup_online'] = dados['backup_online'].map({'Sim' : 1, "Não" : 0})
dados['protecao_dispositivo'] = dados['protecao_dispositivo'].map({'Sim' : 1, "Não" : 0})
dados['suporte_tecnico'] = dados['suporte_tecnico'].map({'Sim' : 1, "Não" : 0})
dados['streaming_TV'] = dados['streaming_TV'].map({'Sim' : 1, "Não" : 0})
dados['streaming_filmes'] = dados['streaming_filmes'].map({'Sim' : 1, "Não" : 0})
dados['fatura_digital'] = dados['fatura_digital'].map({'Sim' : 1, "Não" : 0})
```

Visualizamos o resultado, notando que os rótulos das variável são números.

```python
dados.head()
```

 - coluna **'genero'**



```python
dados['genero_feminino'] = dados['genero'].map({'Feminino' : 1, "Masculino" : 0})
dados = dados.drop('genero', axis=1)
```

- colunas **'streaming_filmes'** e **'streaming_TV'**


```python
dados['servicos_streaming'] = dados['streaming_filmes'] + dados['streaming_TV']
```

```python
dados = dados.drop('streaming_filmes', axis=1)
dados = dados.drop('streaming_TV', axis=1)
```

 - colunas **'backup_online'**/ **'protecao_dispositivo'** / **'suporte_tecnico'** / **'seguranca_online'**

```python
dados['servicos_internet_adicionais'] = dados['backup_online'] + dados['protecao_dispositivo'] + dados['suporte_tecnico'] + dados['seguranca_online']
```

```python
dados = dados.drop('backup_online', axis=1)
dados = dados.drop('seguranca_online', axis=1)
dados = dados.drop('suporte_tecnico', axis=1)
dados = dados.drop('protecao_dispositivo', axis=1)
```

- coluna **'meio_pagamento'**

```python
dados['meio_pagamento']
```

```python
dados['pagamento_manual'] = dados['meio_pagamento'].map({'Cheque eletrônico' : 1,
                                                         'Cheque via correio': 1,
                                                         'Transferência bancária (automático)': 0,
                                                         'Cartão de crédito (automático)': 0})
```

```python
dados = dados.drop('meio_pagamento', axis=1)
```

```python
dados['pagamento_manual']
```

 - colunas **'fatura_mensal'** / **'custo_total'** / **'contas_diarias'**


```python
dados = dados.drop('contas_diarias', axis=1)
dados = dados.drop('custo_total', axis=1)
```

```python
categoricas = ['provedor_internet', 'tipo_contrato']
```

```python
dados = one_hot_encoding.fit_transform(dados)
dados = pd.DataFrame(dados, columns=one_hot_encoding.get_feature_names_out())
```

## 4.  Modelagem

Após as as transformações, fizemos cópias do dataframe 'dados' sendo 'df' e 'df_normal', considerando que 'df_normal' é destinado a ser normalizado.

```python
df = dados.copy()
```

```python
df_normal = df.copy()
```

Prosseguimos com a modelagem dos dados e, para isto, vamos separar os dados:

 -  X:  onde estarão os atributos, features ou variáveis explicativas.
 - y : variável alvo, target ou variável explicada


```python
y = df['cliente_ativo']
X = df.drop(columns='cliente_ativo')
```

- Separação das bases de treino e teste, considerando 30% para teste e 70% para treino do modelo


```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=4652, stratify=y)
```

 - Modelo Baseline | DummyClassifier

```python
from sklearn.dummy import DummyClassifier

modelo1 = DummyClassifier(strategy='most_frequent')
modelo1.fit(X_treino, y_treino)
y_previsto_modelo1 = modelo1.predict(X_teste)
```

* Avaliação do modelo.

```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
modelo1_acuracia = round(accuracy_score(y_teste, y_previsto_modelo1), 4)
modelo1_precisao = round(precision_score(y_teste, y_previsto_modelo1), 4)
modelo1_recall = round(recall_score(y_teste, y_previsto_modelo1), 4)
modelo1_f1 = f1_score(y_teste, y_previsto_modelo1)

y_proba_modelo1 = modelo1.predict_proba(X_teste)[:, 1]
modelo1_roc =  roc_auc_score(y_teste, y_proba_modelo1)
```

```python
lista_modelo1 = [modelo1_acuracia, modelo1_precisao, modelo1_recall, modelo1_f1, modelo1_roc]
nomes = ['Acurácia', 'Precisão', 'Recall', 'F1-Score', 'ROC-AUC']
avaliacao1 = pd.DataFrame({'DummyClassifier': lista_modelo1}, index=nomes)
avaliacao1
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

modelo1_matrix = confusion_matrix(y_teste, y_previsto_modelo1)
print(modelo1_matrix)

disp = ConfusionMatrixDisplay(confusion_matrix=modelo1_matrix, display_labels=None)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusão | DummyClassifier")
plt.show()
```

- Modelo | RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

modelo2 = RandomForestClassifier()
modelo2 = modelo2.fit(X_treino, y_treino)
```

```python
y_previsto_modelo2 = modelo2.predict(X_teste)
```

```python
modelo2_acuracia = round(accuracy_score(y_teste, y_previsto_modelo2), 4)
modelo2_precisao = round(precision_score(y_teste, y_previsto_modelo2), 4)
modelo2_recall = round(recall_score(y_teste, y_previsto_modelo2), 4)
modelo2_f1 = f1_score(y_teste, y_previsto_modelo2)

y_proba_modelo2 = modelo2.predict_proba(X_teste)[:, 1]
modelo2_roc = roc_auc_score(y_teste, y_proba_modelo2)
```

```python
lista_modelo2 = [ modelo2_acuracia, modelo2_precisao, modelo2_recall, modelo2_f1, modelo2_roc]
nomes = [ 'Acurácia', 'Precisão', 'Recall', 'F1-Score', 'ROC-AUC']
avaliacao2 = pd.DataFrame({'RandomForest': lista_modelo2}, index=nomes)
avaliacao2
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
# Calcula a matriz de confusão
modelo2_matrix = confusion_matrix(y_teste, y_previsto_modelo2)
print("Matriz de Confusão (valores brutos):")
print(modelo2_matrix)

# Exibe a matriz de confusão graficamente
disp = ConfusionMatrixDisplay(confusion_matrix=modelo2_matrix, display_labels=None)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusão | RandomForestClassifier")
plt.show()
```

 O modelo de Regressão Logística necessitam da normalização, por serem sensíveis à escala e baseados em cálculo de distância. É importante fazer este procedimento. Para isso, será utilizado o StandardScaler.

```python
y1 = df_normal['cliente_ativo']
X1 = df_normal.drop(columns='cliente_ativo')
```

```python
lista_modelo3 = [ modelo3_acuracia, modelo3_precisao, modelo3_recall, modelo3_f1, modelo3_roc]
nomes = [ 'Acurácia', 'Precisão', 'Recall', 'F1-Score', 'ROC-AUC']
avaliacao3 = pd.DataFrame({'LogisticRegression': lista_modelo3}, index=nomes)
avaliacao3
```
